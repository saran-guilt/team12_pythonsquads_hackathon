```python
#create an array ,np.arange(12).reshape(3,4). 3 is rows and 4 is columns, it is 2D arrays
arr_1 = np.arange(12)
print(arr_1)
reshaped = np.reshape(arr_1,(3,4))
print("reshaped :\n", reshaped)
```

```python
#Replace the array elements with their square values if the element is an even number. Keep the odd numbers as such. 
#Display the resulting array.
arr_convert = np.where(arr_1 % 2 == 0, arr_1 ** 2, arr_1)
arr_result = np.reshape(arr_convert, (3,4))
print("arr_result:\n", arr_result)
```

```python
#Q36 Create a barchart for HR Zone vs count and apply hatches to the bars.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("test_measure.csv")
#df.head(10)

#count HR zone
value_counts = df['HR'].value_counts()

#create bar chart
plt.bar(value_counts.index, value_counts.values, hatch='X', color='pink', edgecolor='black', width=2.9, align='center')

# Add labels and title
plt.xlabel('HR')
plt.ylabel('Counts')
plt.title('Bar Chart of HR zone vs Counts')
plt.show()

#plt.bar(HR, color='pink', edgecolor='red', width=0.5, align='center', hatch='X')
#plt.ylabel('HR')
#plt.title('HR zone vs Count')
```
