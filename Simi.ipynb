```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("test_measure.csv")  

# Melt the dataset for easier plotting (reshape data)
melted_df = df.melt(value_vars=['HR', 'RR'], 
                    var_name='Measurement', 
                    value_name='Value')

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Measurement', y='Value', data=melted_df, palette='muted')

# Add labels and a title
plt.title('Violin Chart for Heart Rate and Respiration Rate')
plt.xlabel('Measurement Type')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# 65. Calculate the percentage of male athletes who exceeded, the maximum treadmill speed of female athletes .

import pandas as pd
subject_info= pd.read_csv("subject-info.csv")
test_measure=pd.read_csv("test_measure.csv")
merge_subject_test= pd.merge(subject_info,test_measure, on='ID')
#max_speed_female=merge_subject_test.loc[merge_subject_test['Sex']=='1','Speed'].max()
max_speed_female = merge_subject_test.loc[merge_subject_test['Sex'] == 1, 'Speed'].max()
print(f"The Maximum Speed of Female is :{max_speed_female} ")

Male_exceeding_femalespeed = merge_subject_test.loc[(merge_subject_test['Sex'] == 0) & 
                                                    (merge_subject_test['Speed'] > max_speed_female)].shape[0]

#Total Male count

male_count=merge_subject_test.loc[merge_subject_test['Sex']==0].shape[0]

print(f"Count of Men {male_count}")

#Calcule the percentage

percent_exceeding=(Male_exceeding_femalespeed/male_count) * 100

print(f"The percentage of male athletes who exceeded the maximum treadmill speed of female athletes is: {percent_exceeding:.2f}%")
```
