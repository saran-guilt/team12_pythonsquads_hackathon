```python
# 5. Create a heatmap to visualize the correlations among any three cardiorespiratory measurements, Age, and BMI.
import pandas as pd 
import matplotlib.pyplot as plot 
import seaborn as sns

df1 = pd.read_csv('subject-info.csv') 
df2 = pd.read_csv('test_measure.csv')



df1['HeightInMeter'] = df1['Height'] / 100
df1['BMI'] = df1['Weight'] / (df1['HeightInMeter'] ** 2) 

merged_df = pd.concat([df1, df2], ignore_index=True)

colms = [ 'Age' , 'BMI' , 'HR' , 'VO2' , 'RR'] 
data_correl = merged_df[colms]
correLation = data_correl.corr()

#for heatmap
plot.figure(figsize = (6,4))
sns.heatmap(correLation, annot=True, cmap='coolwarm', fmt=".2f") 
plot.title('correLation heatmat for Age BMI and cordiorespiratory measurements') 
plot.tight_layout()
plot.show()
```

```python
# 6. Generate subplots for second-degree trigonometric functions,using x=np.arange(0,4*np.pi,0.1) in object oriented approach.Label every π interval on X axis.

import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.arange(0, 4 * np.pi, 0.1)

# Subplots for second-degree trigonometric functions
sin2_x = np.sin(x) ** 2
cos2_x = np.cos(x) ** 2

# using object-oriented approach
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

# Plot sin^2(x)
ax1.plot(x, sin2_x, color='blue')
ax1.set_title(r'$\sin^2(x)$')
ax1.grid(True)

# Plot cos^2(x)
ax2.plot(x, cos2_x, color='green')
ax2.set_title(r'$\cos^2(x)$')
ax2.grid(True)

# Set π interval ticks on x-axis
pi_ticks = np.arange(0, 4 * np.pi + 0.1, np.pi)
pi_labels = [f"{i}π" if i != 0 else "0" for i in range(5)]
ax2.set_xticks(pi_ticks)
ax2.set_xticklabels(pi_labels)

# Add x and y axis labels
ax2.set_xlabel('x (radians)')
ax1.set_ylabel('Amplitude')
ax2.set_ylabel('Amplitude')

# layout
plt.tight_layout()
plt.show()
```

```python
# 9. What percentage of athletes exceeded 90% of max HR after warm up period?

import pandas as pd

df1 = pd.read_csv('subject-info.csv') 
df2 = pd.read_csv('test_measure.csv')

merged_df = pd.concat([df1, df2], ignore_index=True)
max_hr = 220 - merged_df['Age']

def percentage_exceeded_90(merged_df, age_col='Age', warmUpHr='HR'):
    """
    Calculates percentage of athletes exceeding 90% of max HR after warm-up.

    Parameters:
    merged_df (pd.DataFrame): Dataset containing heart rate and age data.
    age_col (float): Column name for age.
    warmup_hr_col (float): Column name for warm-up heart rate.

    Returns:
    float: Percentage of athletes who exceeded 90% of max HR.
    """
    merged_df['max_hr'] = 220 - merged_df[age_col]
    merged_df['HR_90_percent'] = 0.9 * merged_df['max_hr']
    
    count_excd = (merged_df[warmUpHr] > merged_df['HR_90_percent']).sum()
    total = merged_df.shape[0]
      
    percent = (count_excd / total) * 100
    return round(percent, 2)

    # Call the function
result = percentage_exceeded_90(merged_df, age_col='Age', warmUpHr='HR')
print(f"{result}% of athletes exceeded 90% of max HR after warm-up.")
print(merged_df[['Age', 'HR']])
print(f"Max heart rate : ", max_hr)
```

```python
# 14. Calculate RER (Respiratory Exchange Ratio) , and categorize exercise phase to warmup, moderate, peak exercise according to RER
# RER = VCO2/VO2, Warm-up: RER < 0.85, Moderate Exercise: 0.85 ≤ RER < 1.0, Peak Exercise: RER ≥ 1.0
import pandas as pd
df = pd.read_csv('test_measure.csv')

#for RER
df['RER'] = df['VCO2'] / df['VO2']

#exercise phase
def categorize_phase(rer):
    if rer < 0.85 :
        return 'Warm-up'
    elif rer > 0.85 and rer < 1.0:
        return'Moderate Exercise'
    else :
        return'Peak exercise'

df['all_exer_phase'] = df['RER'].apply(categorize_phase)
print(df[['VO2' , 'VCO2', 'RER', 'all_exer_phase']])
```

```python
# 15.Calculate the moving average(rolling mean) of HR for participant with ID 2 in a window size of 10 and plot a graph to show both moving average and original HR.

import pandas as pd
import matplotlib.pyplot as plot

df = pd.read_csv('test_measure.csv')
participant_id = df[df['ID'] == 2].reset_index(drop =True)
participant_id['moving_hr'] = participant_id['HR'].rolling(window=10).mean()

#ploting the calculatiing data
fig = plot.figure(figsize = (6,4))
plot.plot(participant_id['HR'], label = 'Original HR', marker = 'o')
plot.plot(participant_id['moving_hr'], label = 'Moving average', linestyle = '--', color = 'orange')
plot.title('Moving agerage of HR for participant ID is 2')
plot.legend()
plot.grid(True)
plot.tight_layout()
plot.show()
```
