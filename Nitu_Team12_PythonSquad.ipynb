```python
import pandas as pd
from io import StringIO
# 1. What is the average age of Female athletes who participated in this graded exercise tests?
df = pd.read_csv('subject-info.csv')
df.head
female_athletes = df[df['Sex'] == 0]
average_age = female_athletes['Age'].mean()
print(f"Average age of female athletes: {average_age:.2f} years")
```

```python
# 2. Visualize age distribution of all athletes and display the median value in the chart.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('subject-info.csv')
data = {'Age': np.random.randint(15, 45, 1000)}
df = pd.DataFrame(data)

# Calculate the median age
median_age = df['Age'].median()

# Create the histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=False, bins=30)

# Add a vertical line for the median age
plt.axvline(median_age, color='red', linestyle='dashed', linewidth=1, label=f'Median Age: {median_age}')

# Customize the plot
plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.title('Age Distribution of Athletes')
plt.legend()
plt.grid(axis='y', alpha=0.75)

# Display the plot
plt.show()
```

```python
#3. Find the maximum time taken for all tests and then visualize the distribution of number of tests based on the test duration in minutes. 
# Use a line chart and display the count of tests in the chart as well.
import pandas as pd
import matplotlib.pyplot as plot
#df = pd.read_csv('subject-info.csv')
df = pd.read_csv('test_measure.csv')
#print(df.columns)

maxTime = df['time'].max()
print(f"Maximum time taken for all test  {maxTime} minutes")

testDurationMin = df['time'].value_counts().sort_index()
print(f"number of tests based on the test duration in minutes ", testDurationMin)

#line chart and display the count of tests in the chart as well
plot.figure(figsize = (5,3))
plot.plot(testDurationMin.index, testDurationMin.values, marker='o', linestyle ='-')
plot.xlabel('test duration in (minutes)')
plot.ylabel('count of test')
plot.grid('true')
plot.title('display the count of tests in the (munutes)')
plot.tight_layout()
plot.show()
```

```python
# 4. Display the consolidated list of overweight or obese people based on their BMI.
import pandas as pd
df = pd.read_csv('subject-info.csv')

df['HeightInMeter'] = df ['Height'] /100

df ['calculatedBmi'] = df['Weight'] / (df['HeightInMeter']**2)
    
overweight = df[df['calculatedBmi'] >= 25]
print("List of overweight or obese peoples based on their BMI : ")
print( overweight)
```

```python
# 5. Create a heatmap to visualize the correlations among any three cardiorespiratory measurements, Age, and BMI.
import pandas as pd 
import matplotlib.pyplot as plot 
import seaborn as sns

df1 = pd.read_csv('subject-info.csv') 
df2 = pd.read_csv('test_measure.csv')

df1['HeightInMeter'] = df1['Height'] / 100
df1['BMI'] = df1['Weight'] / (df1['HeightInMeter'] ** 2) 

merged_df = pd.concat([df1, df2], ignore_index=True)

colms = [ 'Age' , 'BMI' , 'HR' , 'VO2' , 'RR'] 
data_correl = merged_df[colms]
correLation = data_correl.corr()

#for heatmap
plot.figure(figsize = (6,4))
sns.heatmap(correLation, annot=True, cmap='coolwarm', fmt=".2f") 
plot.title('correLation heatmat for Age BMI and cordiorespiratory measurements') 
plot.tight_layout()
plot.show()
```
